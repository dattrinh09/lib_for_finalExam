```
class Person:
  def __init__(self,first,last):
    self.firstname = first
    self.lastname = last
  

  def describle(self):
    return self.firstname + " " + self.lastname

guido = Person("Guido", "Van Rossum")
print(guido.describle())
```

```
class employee(Person):
  def __init__(self,first,last,staffnum):
    Person.__init__(self,first,last)
    self.staffnum = staffnum
  def describle(self):
    return self.firstname + " " + self.lastname + "," + str(self.staffnum)
guido = employee("gui do","van Rossum",123456)
print(guido.describle()) 
```

```
# y = 0.54 * x - 0.077
import numpy as np

x = np.random.randint(1,73,(100))

#np.random.normal?
y = 0.54 * x - 0.077 + np.random.normal(scale = 1,size = x.shape)

x,y
```

```
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```
c = np.array([np.sum(y),np.sum(y * x)])
A = np.array([[100,np.sum(x)],[np.sum(x),x.T @ x]])

A,c
```

```
alpha , beta = np.linalg.inv(A) @ c
alpha , beta
```

```
plt.scatter(x,y)
x_domain = np.array([0,72])
plt.plot(x_domain,beta * x_domain + alpha)
plt.show
```
